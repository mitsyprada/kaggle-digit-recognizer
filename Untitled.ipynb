```python
img_train_digits = np.empty((28,5628))
col = np.empty((28,28))
row_cnt = 0
px_cnt = 0
for m in range(42000):
    px_cnt+=1
    col = np.hstack((col, train_ar[:,:,m]))
    if(px_cnt%200==0):
        if(row_cnt==0):
            img_train_digits = col
        else:
            img_train_digits = np.vstack((img_train_digits, col))
        row_cnt+=1
        col = np.empty((28,28))

img_train_digits = img_train_digits[:,29:]
```
